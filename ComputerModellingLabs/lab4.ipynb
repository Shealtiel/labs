```python
import numpy as np
import time
import random
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
output_notebook()
```

```python
def random_walk(x, y):
    bound = 50
    j = 0
    t = [time.time()]
    while True:
        vertical = True if random.randint(0,3) != 0 else False

        if vertical: 
            step_x = random.randint(0,1)
            if step_x == 1 and x[j] < bound:
                v = x[j] + 1
            elif x[j] > -bound:
                v = x[j] - 1
            h = y[j]
        else:
            step_y = random.randint(0,1)
            if step_y == 1 and y[j] < bound:
                h = y[j] + 1
            elif y[j] > -bound:
                h = y[j] - 1
            v = x[j]


        x.append(v)
        y.append(h)
        t.append(time.time())
        j += 1
        if x[j] == x[0] and y[j] == y[0]: 
            break

    total_time = t[-1]-t[0]
    return x, y, total_time, j
```

```python
tarray = sarray = []
N = 200

for i in range(N):
    x = [0]
    y = [0]

    _, _, at, asteps = random_walk(x, y)
    tarray = np.append(tarray, at)
    sarray = np.append(sarray, asteps)

print(f'Mean time = {tarray.mean()}')
print(f'Median steps = {np.median(sarray)}')  
```

```python
hist, edges = np.histogram(tarray)
p = figure(title="Distribution")
trace = p.quad(top=hist/N, bottom=0, left=edges[:-1], right=edges[1:])
show(p)
```
